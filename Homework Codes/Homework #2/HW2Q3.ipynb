```python
import numpy as np
from scipy.integrate import solve_ivp
```

```python
def T(t):
    return t*6

def P(t):
    return (t-2)**2 * 8
```

```python
def ode(t, S, a, b, c):
    dydt = P(t) - a * max(T(t), 0) - b * S**c
    return dydt
```

```python
solution = solve_ivp(ode, (0, 5), 8, args=(1, 1, 1), t_eval=np.linspace(0, 5, 100))


```
